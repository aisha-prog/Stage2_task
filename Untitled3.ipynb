```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import linregress
```

```python
#Import both datasets
df_sift = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/sift.csv", sep=',')
print (df_sift.head(5))
print (df_sift.columns)
```

```python
df_foldx = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/foldx.csv", sep=',')
print (df_foldx.head(5))
print (df_foldx.columns)
```

```python
# create a column specific_Protein_aa which will be a cantenation of the Protein and Amino_acid columns such that If you have Protein A5A607 and Amino_acid E63D, you have specific_Protein_aa A5A607_E63D
# This concatenates the 'Protein' and 'Amino_acid' columns with an underscore.
df_sift["specific_protein_aa"] = df_sift["Protein"].astype(str) + "_" + df_sift["Amino_Acid"].astype(str)
print ("SIFT DataFrame:")
print (df_sift.head(5))
```

```python
# This concatenates the 'Protein' and 'Amino_acid' columns with an underscore.
df_foldx["specific_protein_aa"] = df_foldx["Protein"].astype(str) + "_" + df_foldx["Amino_Acid"].astype(str)
print ("FoldX DataFrame:")
print (df_foldx.head(5))
```

```python
# Merge the two DataFrames using the specific_Protein_aa column as the key.
# The 'outer' merge keeps all rows from both datasets.
sift_foldx = pd.merge(df_sift, df_foldx, on="specific_protein_aa", how="outer")
print (sift_foldx.head(5))
```

```python
# Drop NaN rows
sift_foldx.dropna()
```

```python
# Find all mutations that have a SIFT score below 0.05 and FoldX Score above 2 (i.e: Mutations that affect both structure and function)
filtered_sift_foldx = sift_foldx.loc[(sift_foldx["sift_score"] < 0.05) & (sift_foldx["foldx_score"] > 2)]
print (filtered_sift_foldx.head(5))
```

A SIFT Score below 0.05 is deleterious
A FoldX score greater than 2 kCal/mol is deleterious

```python
# Investigate for the amino acid that has the most functional and structural impact
# Using the amino acid column, find a way to select the first amino acid
filtered_sift_foldx['first_amino_acid'] = filtered_sift_foldx['Amino_Acid_x'].astype(str).str[0]
print(filtered_sift_foldx[['Amino_Acid_x', 'first_amino_acid']].head())
```

```python
# Generate a frequency table for all the amino acids
amino_acid_counts = filtered_sift_foldx['first_amino_acid'].value_counts()
print(amino_acid_counts)
```

```python
# Using the amino frequency table above, generate a barplot and pie chart to represent the frequency of the amino acids.
amino_acid_counts.plot(kind='bar', title='Amino Acid Frequency')
plt.xlabel('Amino Acid')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.figure(figsize=(8,8))
plt.pie(amino_acid_counts, labels=amino_acid_counts.index, autopct='%1.1f%%', )
plt.title('Amino Acid Frequency')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import linregress
```

```python
df_genefold = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/results.csv", sep=',')
print (df_genefold.head(5))
print (df_genefold.columns)
```

```python
#Adjust pvalue to the -log10 scale
df_genefold['neg_log10_pvalue'] = -np.log10(df_genefold['pvalue'])
```

```python
df_genefold['color'] = 'gray'  # Default color
df_genefold.loc[(df_genefold['log2FoldChange'] > 1) & (df_genefold['pvalue'] < 0.01), 'color'] = 'red'  # Upregulated
df_genefold.loc[(df_genefold['log2FoldChange'] < -1) & (df_genefold['pvalue'] < 0.01), 'color'] = 'blue'  # Downregulated
```

```python
# Generate a volcano plot using scatter plot
plt.figure(figsize=(8,6))
plt.scatter(df_genefold['log2FoldChange'], df_genefold['neg_log10_pvalue'], c=df_genefold['color'], alpha=0.5)

# Add horizontal line for p-value threshold (0.01)
plt.axhline(y=-np.log10(0.01), color='green', linestyle='--')

# Add vertical lines for log2FoldChange thresholds (-1 and 1)
plt.axvline(x=1, color='green', linestyle='--')
plt.axvline(x=-1, color='green', linestyle='--')

# Labeling the axes and the plot
plt.xlabel("log2FoldChange")
plt.ylabel("pvalue")
plt.title("Volcano Plot")

plt.show()
```

A volcano plot is a scatter plot that shows the statistical significance of changes in data. It's a common way to display data in biological research.
How it works
The x-axis shows the fold change, or the magnitude of change in the data
The y-axis shows the adjusted p-value, or the statistical significance of the change
Points with low p-values (i.e. highly significant) appear toward the top of the plot
Points with large fold changes appear toward the right of the plot
Why it's useful
Volcano plots are useful for quickly identifying genes that have large changes and are statistically significant. These genes may be the most biologically significant.
When it's used
Volcano plots are commonly used in omic experiments such as genomics, transcriptomics, proteomics, and metabolomics.
A volcano plot is constructed by plotting the negative logarithm of the p value on the y axis (usually base 10)

```python
# Determine the upregulated genes (Genes with Log2FC > 1 and pvalue < 0.01)
upregulated_genes = df_genefold[(df_genefold['log2FoldChange'] > 1) & (df_genefold['pvalue'] < 0.01)]
print (upregulated_genes.head())

# Determine the downregulated genes (Genes with Log2FC < -1 and pvalue < 0.01)
downregulated_genes = df_genefold[(df_genefold['log2FoldChange'] < -1) & (df_genefold['pvalue'] < 0.01)]
print (downregulated_genes.head())
```

```python
# Conduct t-tests
# Age and Gender
age_male = df_nhanes[df_nhanes['Gender'] == 'Male']['Age']
age_female = df_nhanes[df_nhanes['Gender'] == 'Female']['Age']
t_stat, p_value = stats.ttest_ind(age_male, age_female, nan_policy='omit') # Conduct t-test and specify handling of NaNs
print(f"T-test between Age and Gender: t-statistic = {t_stat}, p-value = {p_value}")

# BMI and Diabetes
bmi_diabetic = df_nhanes[df_nhanes['Diabetes'] == 'Yes']['BMI']
bmi_non_diabetic = df_nhanes[df_nhanes['Diabetes'] == 'No']['BMI']
t_stat, p_value = stats.ttest_ind(bmi_diabetic, bmi_non_diabetic, nan_policy='omit') # Conduct t-test and specify handling of NaNs
print(f"T-test between BMI and Diabetes: t-statistic = {t_stat}, p-value = {p_value}")

# Alcohol Year and Relationship Status
alcohol_year_married = df_nhanes[df_nhanes['RelationshipStatus'] == 'Married']['AlcoholYear']
alcohol_year_single = df_nhanes[df_nhanes['RelationshipStatus'] != 'Married']['AlcoholYear']
t_stat, p_value = stats.ttest_ind(alcohol_year_married, alcohol_year_single, nan_policy='omit') # Conduct t-test and specify handling of NaNs
print(f"T-test between Alcohol Year and Relationship Status: t-statistic = {t_stat}, p-value = {p_value}")
```
